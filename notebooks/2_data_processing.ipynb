```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

Concatenate the dataframes of train.csv and test.csv

```python
df_train = pd.read_csv('../data/row/train.csv')
df_test = pd.read_csv('../data/row/test.csv')

df = pd.concat([df_train, df_test], ignore_index=True)

df
```

Dropping the 'Name' column

```python
df.drop('Name', axis='columns', inplace=True)

df
```

Breaking the 'Cabin' column into 3 parts using the separtor "/" and deleting the 'CabinNumber' column as it is not needed

```python
df[['CabinDeck', 'CabinNumber', 'CabinSide']] = df.Cabin.str.split(pat='/', expand=True)

df.drop(['Cabin', 'CabinNumber'], axis=1, inplace=True)

df
```

Separate the column 'PassengerId' into 'PassengerGroup' and 'PassengerGroupId' with '_' separator, and then delete the 'PassengerGroupId' and 'PassengerId'

```python
df[['PassengerGroup', 'PassengerGroupNumber']] = df.PassengerId.str.split(pat='_', expand=True)
df.drop(['PassengerId', 'PassengerGroupNumber'], axis=1, inplace=True)

df
```

Summing all the 5 spent columns into a new column 'MoneySpent' and then deleting these 5

```python
columns_to_remove = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']

df['MoneySpent'] = df[columns_to_remove].sum(axis=1)

df.drop(columns=columns_to_remove, inplace=True)

df
```

Create a new column 'CryoSleepValueAvail' and add TRUE if 'CryoSleep' value is available otherwise FALSE

```python
df['CryoSleepValueAvail'] = df['CryoSleep'].notna()

df
```

Checking if the above formula has worked

```python
df[df['CryoSleep'].isna()]
```

Fill Missing Values with FALSE in columns 'CryoSleep' and 'VIP', pd.set_option() is to avoid FutureWarning and astype(bool) is to make sure the filled values are in bool type

```python
pd.set_option('future.no_silent_downcasting', True)

df[['CryoSleep', 'VIP']] = df[['CryoSleep', 'VIP']].fillna(False).astype(bool)

df
```

Add a new column 'HomePlanetValueAvail' and add TRUE if 'HomePlanet' value is available otherwise FALSE, also fill missing values of 'HomePlanet' with 'Europa'

```python
df['HomePlanetValueAvail'] = df.HomePlanet.notna()
df.HomePlanet = df.HomePlanet.fillna('Europa').astype(str)

df
```

Create a new column 'CabinValueAvail' and if value in 'CabinSide' available then TRUE otherwise FALSE

```python
df['CabinValueAvail'] = df.CabinSide.notna()
df
```

```python
df.to_csv('../data/processed/before_cabin_fillna.csv', index=False)
```

We have 3 columns in the df, 'PassengerGroup', 'CabinDeck' and 'CabinSide', there are some missing values in 'CabinDeck' and 'CabinSide' at the same time, and there are some repeating values in 'PassengerGroup', I have assumed that (from the currrent data), the people who have the same 'PassengerGroup' have the higher probability of haveing the same 'CabinDeck' bzw. 'CabinSide' I am filling up the 'CabinDeck' and 'CabinSide' values if the 'PassengerGroup' value is same(duplicated), so for example if 'PassengerGroup': 0003 and 'CabinDeck': A and 'CabinSide': S, then for another occurrence of 0003 should also have 'CabinDeck': A and 'CabinSide': S, where there is no repetition of 'PassengerId', then I am putting 'F' in 'CabinDeck' and 'S' in 'CabinSide', because these values are have most repeatation bzw. Mode

```python
def fill_missing_values_from_PassengerGroup_to_Cabin(series: pd.Series) -> pd.Series:
    return series.ffill().bfill()
```

```python
df.CabinDeck = df.groupby('PassengerGroup')['CabinDeck'].transform(fill_missing_values_from_PassengerGroup_to_Cabin)
df.CabinSide = df.groupby('PassengerGroup')['CabinSide'].transform(fill_missing_values_from_PassengerGroup_to_Cabin)

df.CabinDeck = df.CabinDeck.fillna('F')
df.CabinSide = df.CabinSide.fillna('S')

df
```

```python
df.to_csv('../data/processed/after_cabin_fillna.csv', index=False)
```

```python
df.drop('PassengerGroup', axis=1, inplace=True)

df
```

```python
df.Destination = df.Destination.fillna('TRAPPIST-1e').astype(str)

df
```

Scaling the values of Age in the range of [0, 1]

```python
scaler = MinMaxScaler()

df[['Age', 'MoneySpent']] = scaler.fit_transform(df[['Age', 'MoneySpent']])

df
```

Convert the TRUE into 1 and FALSE into 0 for columns 'CryoSleep', 'VIP', 'Transported', 'CryoSleepValueAvail', 'HomePlanetValueAvail', 'CabinValueAvail', and for NaN keep NaN (for our Target Column)

```python
columns_to_convert = ['CryoSleep', 'VIP', 'Transported', 'CryoSleepValueAvail', 'HomePlanetValueAvail', 'CabinValueAvail']

df[columns_to_convert] = df[columns_to_convert].map(lambda x: 1 if x is True else (0 if x is False else x))

df
```

Converting categorical columns ['HomePlanet', 'Destination', 'CabinDeck', 'CabinSide'] into numeric data using OneHotEncoder

```python
categorical_columns = ['HomePlanet', 'Destination', 'CabinDeck', 'CabinSide']

one_hot_encoder = OneHotEncoder(sparse_output=False, drop='first')
transformer = ColumnTransformer([('one_hot', one_hot_encoder, categorical_columns)], remainder="passthrough")

transformed_array = transformer.fit_transform(df)

df_transformed = pd.DataFrame(transformed_array)

df_transformed
```
