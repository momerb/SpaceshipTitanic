```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/graph.csv')

df
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['MoneySpent'], df['Age'], color='blue', alpha=0.7)
plt.title('Comparison of Age and Money Spent')
plt.xlabel('Age')
plt.ylabel('Money Spent')
plt.grid(True)
plt.show()
```

The graph shows that age < 14 has 0 spent, so this information can be used to predict the missing values of 'MoneySpent' atleast for the age less than 14, for age greater than 14, we will have to take the Median.
